# Pneumonia Detection Using Chest X-Rays
### Applied Supervised Machine Learning Using Python


__About the Dataset__

The Chest X Ray images were imported from Kaggle using the Kaggle API. 
https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia?resource=download

_Citation: Kermany, Daniel; Zhang, Kang; Goldbaum, Michael (2018), "Labeled Optical Coherence Tomography (OCT) and Chest X-Ray Images for Classification", Mendeley Data, v2._

The dataset is organized into 3 folders (train, test, val) and contains subfolders for each image category (Pneumonia/Normal). There are 5,863 X-Ray images (JPEG) and 2 categories (Pneumonia/Normal).

Chest X-ray images (anterior-posterior) were selected from retrospective cohorts of pediatric patients of one to five years old from Guangzhou Women and Children’s Medical Center, Guangzhou. All chest X-ray imaging was performed as part of patients’ routine clinical care.

__Interacting with Dataset using the Kaggle API and CLI tool__

https://www.kaggle.com/docs/api

The easiest way to interact with Kaggle’s public API is via their command-line tool (CLI) implemented in Python. See the steps below to see how to interact with the API to download the dataset locally.

1. First I used pip install kaggle. 


2. Next I needed to make an API token. I logged into my Kaggle account, then went to the 'Account' tab of my user profile. From there, I selected 'Create API Token'. This triggered the download of kaggle.json, a file containing my API credentials. 


3. I place this file in the location ~/.kaggle/kaggle.json (where ~ is just C:/users/your_username)


4. I defined a shell environment variable KAGGLE_CONFIG_DIR to change this location to ~.kaggle\kaggle.json.


5. Next, I ran the following command from the command prompt:    

    _kaggle datasets download -d paultimothymooney/chest-xray-pneumonia_

    This downloaded the dataset to my folder C:/users/missm.
